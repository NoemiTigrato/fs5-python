**[Task 1.2 (d).]()** Binarize the image by applying an intensity threshold:

```python
thres = 0.1
imagebin = image >= thres
#es wird durch jedes pixel gegangen und geschaut ob die Bedingung zutrifft, falls ja dann wird TURE angenommen und falls Nein dann FALSE, 
# TRUE und FALSE können dann folgendermaßen angepasst werden TURE = 1 und FALSE =0 und dann übersetzt in Intensitätswerten eben 255 oder Intensität 0 
plt.imshow(imagebin, cmap = 'gray')
```

**[Task 1.2 (e).]()** Load the ground truth:

```python
gt = plt.imread('data/NIH3T3/gt/0.png')
plt.imshow(gt, cmap='gray')
```

Compute and print the Dice coefficient:

```python
G = gt
H = imagebin

D0 = 2*G*H
D1 = D0.sum()
D2 = G.sum()+H.sum()

dice = D1/D2
dice
```

```python
def Dicefun(G,H):
    D0 = 2*G*H
    D1 = D0.sum()
    D2 = G.sum()+H.sum()
    return D1/D2

#um eine funktion zu schreiben muss ich einfach nur def einen Namen und ihm dann sagen welche Elemente er holen muss 
#am Ende return schreiben 
```

```python
Dicefun(gt,imagebin)
```
